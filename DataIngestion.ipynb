```python
import re
```

```python

def remove_handle_tags(text):
    pattern=re.compile('[@|#][a-zA-Z0-9_]*')
    #print(text+"\n")
    return pattern.sub("",text)
```

```python
def remove_url(text):
    #text = text = "@@Tesco Haddington: 6 x Mini Melton Mowbray Pies @ £1.00; 4 x Mini Melton Mowbray Pies on reduced shelf @ £1.04! #TesCon @Tesco  https://ts.co/P7XCmTzPQj"
    pattern=re.compile(r'https?://\S+|[ts|www]\.\S+')
    return pattern.sub("",text)
```

```python
def remove_price(text):
    #text = text = "@@Tesco Haddington: 6 x Mini Melton Mowbray Pies @ £1.00; 4 x Mini Melton Mowbray Pies on reduced shelf @ £1.04! #TesCon @Tesco http://www.google"
    pattern=re.compile('[\d\.\d]*|€|$|£|¥|₹')
    return pattern.sub("",text)
```

```python
def remove_captag(text):
    #print(text)
    text = fr'{text}'
    #print(rf'{text}')
    pattern=re.compile(r'[ *][\r\n]*[\^|-][a-zA-Z0-9_]*')
     
    return pattern.sub("",text)
```

```python
def remove_punc(text):
    exclude=string.punctuation
    exclude+="‘"
    exclude+="’"

    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
def remove_punc(text):
    exclude=string.punctuation
    exclude+="‘"
    exclude+="’"

    for char in exclude:
        text=text.replace(char,"")
    return text
```

```python
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', text)
    return clean_text

```

```python
def remove_html_tags(text):
    pattern=re.compile('<.*?>')
    return pattern.sub("",text)
```

```python
def remove_stopwords(text):
  
  new_text=[]
  for word in text.lower().split():
    if word not in stopwords.words('english'):
      new_text.append(word)
  return " ".join(new_text)
```

```python
def remove_unwanted_strings(text): 
  unwanted=['\u200d']
  new_text=[]
  for word in text.lower().split():
    if word not in unwanted:
      new_text.sub("",text)
  return " ".join(new_text)
```

```python
def generateFinalSet():
    df = pd.DataFrame()
    dataframevalue = []
    for i in range(0,1000):
        #print(data["text"][i])
        text= rf'{data["text"][i]}'
        #print(text)
        filter_text = remove_handle_tags(text)
        filter_text = remove_url(filter_text)
        filter_text = remove_price(filter_text)
        filter_text = remove_captag(filter_text)
        filter_text = remove_emojis_manually(filter_text)
        filter_text = remove_html_tags(filter_text) 
        filter_text = remove_stopwords(filter_text)
        filter_text = remove_punc(filter_text) 
        #print(filter_text)  
        final_string = " ".join(str(element) for element in word_tokenize(filter_text))  
        dataframevalue.append([final_string,data["text"][i]])
        #print("======")
        df = pd.DataFrame(dataframevalue,columns=['Filter', 'Orignal'])
    return df
 
```

```python
filterresult = generateFinalSet() 
```
